## Background

### Landsat Collection 1
Landsat Collection 1 consists of Level 1 products generated from the Landsat 8 Operational Land Imager (OLI) / Thermal Infrared Sensor (TIRS), Landsat 7 Enhanced Thematic Mapper Plus (ETM+), Landsat 4-5 Thematic Mapper (TM), and Landsat 1-5 Multispectral Scanner (MSS) instruments. The implementation of collections ensures consistent and known radiometric and geometric quality through time and across instruments and improves control in the calibration and processing parameters.
Landsat Collection 1 Level 1 products are assigned to one of three Landsat collection tiers:
* Real-Time (RT) 
> OLI / TIRS and ETM+ data processed immediately upon downlink but use predicted ephemeris, initial bumper mode parameters, or
initial TIRS Line of Sight (LOS) model parameters.
* Tier 1
> OLI / TIRS, ETM+, TM, and MSS data processed to Level 1 Precision and Terrain corrected products (L1TP) with image-to-image registration to the GLS control of ≤12-meter Radial Root Mean Square Error (RMSE).
* Tier 2
> OLI / TIRS and ETM+ data processed to Level 1 Systematic and Terrain Corrected products (L1GT), TM and MSS data processed to Level 1 Systematically corrected (L1GS) products, and Level 1 Precision and Terrain Corrected products (L1TP) with image-to-image registration to the GLS control of >12-meter RMSE.

>Product names: "ls8_usgs_sr_scene", "ls7_usgs_sr_scene", "ls5_usgs_sr_scene"

The Processing Level designations for Landsat Collection 1 are shown here. Precision (ground control) and terrain corrected products are labeled with the L1TP designation. Terrain and systematic corrected products are labeled as L1GT. Systematically corrected products are labeled as L1GS. OLI / TIRS products are either L1TP or L1GT and ETM+, TM, and MSS products are L1TP, L1GT, or L1GS.

## Landsat Collection 2

Landsat collection 2 data has a number of processing, geometric, and radiometric improvements, along with a new data distribution process compared to Collection 1. Collection 2 has two levels of product. Level-1 and Level-2.
Landsat Level-2 products are generated from Collection 2 Level-1 inputs that meet the <76 degrees Solar Zenith Angle constraint and include the required auxiliary data inputs to generate a scientifically viable product.

Landsat 8 Operational Land Imager (OLI) surface reflectance products are generated using the Land Surface Reflectance Code (LaSRC) algorithm (Version 1.4.1). Landsat 4-5 Thematic Mapper (TM) and Landsat 7 Enhanced Thematic Mapper (ETM+) surface reflectance products are generated using the Landsat Ecosystem Disturbance Adaptive Processing System (LEDAPS) algorithm (Version 3.4.0).

Landsat 4-8 surface temperature products are generated using the Landsat surface temperature algorithm (Version 1.3.0), developed in cooperation with the Rochester Institute of Technology and NASA Jet Propulsion Laboratory.
A few number of improvements can be envisaged in Collection 2 compared to collection 1:

* Geometric Accuracy
> The Landsat 8 Operational Land Imager (OLI) Ground Control Points (GCPs) is re-baselined to the European Space Agency Copernicus Sentinel-2 Global Reference Image (GRI) to improve the interoperability of the global Landsat archive spatially and temporally.
* Digital Elevation Modeling
> Collection 2 uses the 3-arc-second digital elevation modeling (DEM) sources. 
* Global Level-2 Science and Atmospheric Auxiliary Products
> Processing and distribution of Level-2 surface reflectance and surface temperature science products for Landsats 4-8. Landsat Collection 2 Atmospheric Auxiliary Data used in the processing of Collection 2 Level-2 products are also available to download.
* Consistent Quality Assessment Bands
> Collection 2 Level-1 products are delivered with two additioanl bands called : Quality Assessment Band (QA_Pixel) and a Radiometric Saturation and Terrain Occlusion Quality Assessment Band (QA_RADSAT).
* Updated and Consistent Metadata Files
> In order to facilitate consistency, machine-to-machine scripting and rapid querying of the USGS Landsat archive, two metadata files are included in Collection 2:
Object Description Language-based format called Material Library File(MTL) and Extensive Markup Language(XML)
* Cloud Optimized File Fomat
> Landsat Collection 2 products are delivered in a Cloud Optimized Georeferenced (COG) Tagged Image File Format that is an extension of the current GeoTIFF file format. The COG format improves access to Geospatial datasets in a cloud-based environment by allowing users to request only the data band files that they need. 

## Data Access and Availability 
Landsat Collection 2 data processing will start in 2020, with public availability soon after. The collection will be available to download from USGS [EarthExplorer](https://earthexplorer.usgs.gov/) and LandsatLook. Users are encouraged to reference the [COG Data Format Control Book (DFCB)](https://www.usgs.gov/media/files/landsat-cloud-optimized-geotiff-data-format-control-book) for a detailed description of the image, product and metadata files. Together with the instrument specific Level-1 DFCBs, the COG DFCB provides the user with a comprehensive overview of the new Landsat Collection 2 products.

Level-2 Multi Spectral Scanner (MSS) surface reflectance products will only become available upon the maturity of an operational atmospheric compensation algorithm which satisfies the USGS Landsat program requirements. Collection 1 based products, U.S. Landsat ARD and Level-3 data processing and downloads will remain available for at least one year after Collection 2 is publicly released.

## Surface Reflectance and Surface Temperature
Surface reflectance as an scalar measure shows the fraction of incoming solar radiation that is reflected from the Earth's surface to the Landsat sensor. The LEDAPS and LaSRC surface reflectance algorithms correct for the temporally, spatially and spectrally varying scattering and absorbing effects of atmospheric gases and aerosols, which is necessary to reliably characterize the Earth’s land surface.

Surface Temperature measures the Earth’s surface temperature in Kelvin and is an important geophysical parameter in global energy balance studies, hydrologic modeling, monitoring crop and vegetation health, as well as natural disasters (e.g., volcanic eruptions, wildfires), and urban heat island effects . A surface temperature product cannot be generated if auxiliary data is missing, although a surface reflectance product will continue to be generated.
